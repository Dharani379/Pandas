```python
import pandas as pd
```

## Read to a csv file

```python
df=pd.read_csv('Data/survey_results_public.csv',index_col='Respondent')
```

```python
df.head()
```

## Create a filter

```python
filt = (df['Country'] == 'India')
India_df = df.loc[filt]
India_df.head()
```

## Write to a CSV file

```python
India_df.to_csv('Data/modified_data.csv')
```

## wite to a TSV file

```python
India_df.to_csv('Data/modified_data.tsv', sep='\t')
```

## Read to a TSV file

```python
india_df = pd.read_csv('Data/modified_data.tsv', sep='\t')
india_df
```

## Excel files

```python
India_df.to_excel('Data/modified.xlsx')
```

```python
indiaDf = pd.read_excel('Data/modified.xlsx', index_col = 'Respondent')
indiaDf.head(5)
```
