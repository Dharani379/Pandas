```python
Totalrespondents= df['Country'].value_counts()
Totalrespondents
```

```python
Num_knows_Python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
Num_knows_Python
```

```python
Python_df = pd.concat([Totalrespondents,Num_knows_Python],axis = 'columns',sort=False)
Python_df
```

```python
Python_df.rename(columns={'count':'NumRespondents', 'LanguageWorkedWith':'NumknowsPython'},inplace=True)
Python_df
```

```python
Python_df['PctknowsPython'] = (Python_df['NumknowsPython']/Python_df['NumRespondents'])*100
Python_df['PctknowsPython']
```

```python
Python_df
```

```python
Python_df.sort_values(by='PctknowsPython', ascending=False, inplace=True)
Python_df
```
