```python
import pandas as pd
import numpy as np
```

```python
people = { 
    'first': ['Dharani','Monika','Chinnu','Corey','Hemansh','Jane','Chris'],
    'last' : ['Meka','Meka', np.nan,'Schafer',np.nan,'Doe','Missing'],
    'email' : ['mekadharani@gmail.com','monikameka@2mail.com',None,'coreyschafer@email.com',np.nan,'janedoe@gmail.com','Missing'],
    'age' : ['23','24','25','18','34','66','Missing']
    }
```

```python
df = pd.DataFrame(people)
```

```python
df
```

## Dropping Nan values

```python
df.dropna()
```

```python
df.dropna(axis='index',how='any')
```

```python
df.dropna(axis='columns',how='any')
```

```python
df.dropna(axis='index',how='all')
```

```python
df.dropna(axis='columns',how='all')
```

## Converting to Nan values

```python
df = pd.DataFrame(people)
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
df.dropna(axis='index',how='all',subset=['email'])
```
